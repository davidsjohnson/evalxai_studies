```python
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt

rng = np.random.default_rng()
```

```python
test = stats.ttest_ind
nobs_x = np.arange(10, 100)
nobs_y = nobs_x
n_observations = (nobs_x, nobs_y)

rvs1 = lambda size: rng.normal(0.55, .15, size=size)
rvs2 = lambda size: rng.normal(0.45, .15, size=size)
rvs = (rvs1, rvs2)

res = stats.power(test, rvs, n_observations, significance=0.05)
```

```python
ax = plt.subplot()
ax.plot(nobs_x, res.power)
ax.set_xlabel('Sample Size')
ax.set_ylabel('Simulated Power')
ax.set_title('Simulated Power of t-test')
```

```python
n_observations = (25, 25)
t_observations = sum(n_observations)

loc = np.linspace(0, 1, 20)
rvs2 = lambda size, loc: rng.normal(loc=loc, scale=.15, size=size)
rvs = (rvs1, rvs2)
res = stats.power(test, rvs, n_observations, significance=0.5, kwargs={'loc': loc})
```

```python
ax = plt.subplot()
ax.plot(loc-0.55, res.power) # 0.55 = loc of main distribution
ax.set_xlabel('Effect Size')
ax.set_ylabel('Simulated Power')
ax.set_title(f'Simulated Power of t-test, with varying Effect - {t_observations} Total Samples')
```

```python
# Final

t_observations = 50
n_observations = (t_observations / 2, t_observations / 2)

rvs1 = lambda size: rng.normal(0.55, .10, size=size)
rvs2 = lambda size: rng.normal(0.50, .10, size=size)
rvs = (rvs1, rvs2)

res = stats.power(test, rvs, n_observations, significance=0.05)
print(n_observations, res)
```
