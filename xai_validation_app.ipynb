```python
# prompt: write code to open a list of jpegs
import os
import random
from pathlib import Path
from PIL import Image

random.seed(0)

import pandas as pd

import ipywidgets as widgets
from IPython.display import display
```

```python
gdrive = Path('/content/drive/MyDrive')
```

```python
# prompt: write a function to display one of the images a include a jupyter single select widget not a dropdown.
# The select should have the options: Has Cancer, Does Not Have Cancer, and Not Sure

results = {}
selects = {}

def display_image_with_selector(image, example_num):
  html = widgets.HTML(
    value=f"<div style='padding-top: 20px;'></div><h1>Image Diagnosis - Example {example_num+1}</h1>",
  )
  display(html)

  id, true, pred, _ = Path(image.filename).stem.split( '_')
  true = true.split('=')[-1]
  pred = pred.split('=')[-1]

  opts_map = {'Does Not Have Cancer': 0, 'Has Cancer': 1, 'Not Sure': -1}
  opts = list(opts_map.keys())
  random.shuffle(opts)
  selector = widgets.ToggleButtons(
      options=opts,
      description='Diagnosis:',
      disabled=False
  )
  selector.style.button_color = 'lightblue'
  selects[id] = selector

  results[id] = dict(
      select=opts_map[selector.value],
      true=true,
      pred=pred
  )

  def on_change(change):
    if change['type'] == 'change' and change['name'] == 'value':
      results[id]['select'] = opts_map[change['new']]

  selector.observe(on_change)
  display(image)  # Display the first image initially
  display(selector)
```

```python
image_folder = gdrive / 'hcxai/blocky_diagnosis/original_input/biased_model/2024-07-18'
# ensure order before shuffle for replicability
image_paths = [p for p in image_folder.rglob('*.png')]
image_paths.sort()
images = [Image.open(os.path.join(image_folder, f)) for f in image_paths]
random.shuffle(images)
```

```python
# Generate Selects for all images
for i, image in enumerate(images):
  display_image_with_selector(image, i)

# Create a submit button
html = widgets.HTML(
  value=f"<div style='padding-top: 20px;'></div><h1>Submit Your Diagnoses</h1>",
)
display(html)
submit_button = widgets.Button(description="Submit",
                               icon='check')
submit_button.style.button_color = 'lightblue'
output = widgets.Output()

def on_button_clicked(b):
  with output:
    df = pd.DataFrame.from_dict(results, orient='index').reset_index().rename(columns={'index': 'id'})
    df.to_csv('results.csv', index=False)
    n_corr = (df['true'].astype(int) == df['select'].astype(int)).sum()
    print(f'You got {n_corr}/{len(df)} correct')
    submit_button.disabled = True
    submit_button.description = 'Results Saved'
    submit_button.button_style = 'success'
    for id, selector in selects.items():
      if int(results[id]['select']) == int(results[id]['true']):
        selector.button_style = 'success'
      else:
        selector.button_style = 'danger'

      selector.disabled = True

submit_button.on_click(on_button_clicked)
display(submit_button, output)
```
